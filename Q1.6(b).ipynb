```python
import numpy as np                               # import numpy library as np
import matplotlib.pyplot as plt                  # import mathpltlib.pyplot library as plt
import cmath                                     # import cmath library 

j = cmath.sqrt(-1)                               # iota as j = sqrt(-1)
w = np.linspace(-np.pi, np.pi, 101)              # create a space of w from -pi to + pi

numer = 1 + 3*np.exp(-j*w) + 3*np.exp(-j*2*w) + np.exp(-j*3*w)
denom = 0.0736 - 0.0718*np.exp(-j*w) + 0.0631*np.exp(-j*2*w) + 0.0214*np.exp(-j*3*w)

# X = np.zeros(len(w), dtype = 'complex')
G = numer/denom
# Real and Imaginary parts 
plt.plot(w,G.real,'g')
plt.xlabel("w")
plt.ylabel("Re( Y[n] )")
plt.title("REAL PART PLOT OF DTFT ") 
plt.figure()

plt.plot(w,G.imag, 'r')
plt.xlabel("w")
plt.ylabel("Imag( Y[n] )")
plt.title("IMAG PART PLOT OF DTFT ") 
plt.figure()
```

```python
#phase and magnitude plot 
phaseG = np.zeros([len(w)])              #define total length of phase 
for i in range(0,len(w)):
    phaseG[i] = cmath.phase(G[i])        #phase of Y[n] using cmath library.
    
# Magnitude
plt.plot(w, np.abs(G),'g')               # green color plot of Abs(Y[n]) vs w
plt.xlabel("w")
plt.ylabel("Mag ( G[n] )")
plt.title("Magnitude PLOT ") 
plt.figure()
# Phase
plt.plot(w, phaseG*180/np.pi,'orange')   # green color plot of Phase(Y[n]) in degrees vs w 
plt.xlabel("w")
plt.ylabel("Phase ( G[n] )")
plt.title("Phase angle PLOT ") 
```

```python
plt.magnitude_spectrum(G)
plt.title('Magnitude Spectrum using matplotlib')
plt.figure()
plt.phase_spectrum(G)
plt.title('Phase Spectrum using matplotlib')
```
