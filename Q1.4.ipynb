
# De-convolution of two signals y[n] and h[n]

```python
import numpy as np                               # import numpy library as np
import matplotlib.pyplot as plt                  # import mathpltlib.pyplot library as plt

y=np.array([2,8,20,40,60,68,62,40])           #x[n] 
h=np.array([2,4,6,8])                         #h[n]

Y=len(y)                                      #length of y as Y                                     
H=len(h)                                      #length of h as H
N=Y-H+1
x=np.zeros(N)
sum=0                                         #initialize sum as 0.

skip = np.abs(N - H)

for n in range(0,N):
    sum = 0               #Y= X+H-1 hence X= Y-H+1
    
    for k in range(0,n):                 #range of k from the proven erquation            
        if n==H:                         #if we have same n=N we have to skip the k value 
            k=k+skip                        #skip k with addidng +1
        sum= sum + x[k]*h[n-k]           # from the formula
    x[n]=(y[n]-sum)/h[0]        # generation of de-onvolved x[n]   
    
    
x
```

```python
print(x)
```

```python
import numpy as np                               # import numpy library as np
import matplotlib.pyplot as plt                  # import mathpltlib.pyplot library as plt

y=np.array([-3,8,-20,10,-12,12,16,-58,35])           #x[n] 
h=np.array([-1,2,-4,-4,5])                         #h[n]

Y=len(y)                                      #length of y as Y                                     
H=len(h)                                      #length of h as H
N=Y-H+1
x=np.zeros(N)                                         #initialize sum as 0.


for n in range(N):                       #Y= X+H-1 hence X= Y-H+1
    sum = 0             
    for k in range(n):                   #range of k from the proven erquation            
        if n==N:                         #if we have same n=N we have to skip the k value 
            k=k+1                        #skip k with addidng +1
        sum= sum + x[k]*h[n-k]           # from the formula
    x[n]=(y[n]-sum)/h[0]                 # generation of de-onvolved x[n]       
```
