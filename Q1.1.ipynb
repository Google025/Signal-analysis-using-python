```python
import numpy as np                               #import numpy library as np
import matplotlib.pyplot as plt                  #import mathpltlib.pyplot library as plt
n = np.linspace(0,100,101)                       #creating a space of from 0 to 100 i.e. 101 with 101 samples i.e. 1 unit b/w each point
print(len(n))                                    #check length of n

#generate uncorrupted signal S[n]
s = 2*n*0.9**n                                   #Generating s[n] as given in question                             
plt.plot(n,s)                                    #plot s[n] vs n plot 
plt.xlabel("n")
plt.ylabel("S[n]")
plt.title("Signal S[n]") 
plt.figure()

#generate corrupted noisy random signal d[n]    
d = np.random.normal(0.2,0.05,101)               #generate a random signal as distortion signal 

x = s+d                                          #adding distortion to main signal to get noisy signal.
plt.plot(n,x)                                    #plot of noisy signal x[n] vs n 
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Signal X[n]") 
plt.figure()

v = np.zeros([101])                              #definig a array 'v' to perform 3 point Moving average system
v[0] = x[0]                                      #for 1 size array average will be it self.
v[1] = (x[0] + x[1])/2                           #for 2 size array average will be x[0]+x[1]/2

               #for average of all other inputs average wil be x[n-2]+x[n-1]+x[n]/3 as system is 3 point average system. 
               # y[0] = x[0]
               # y[2] = (x[2]+x[1]+x[0])/3
               # y[3] = (x[3]+x[2]+x[21])/3
               # ...
               # y[100] = (x[100]+x[99]+x[98])/3
               # hence from ssequence the range of average will be from 2 to 101.
for i in range(2,101):                         #for loop to make average of 3 points                           
    v[i] = (x[i]+x[i-1]+x[i-2])/3

print(len(v))

plt.plot(n,v)
plt.xlabel("n")
plt.ylabel("y[n]")
plt.title("Signal Y[n]") 
```
