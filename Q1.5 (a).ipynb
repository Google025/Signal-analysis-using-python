
# Discrete time Fourier transform

```python
import numpy as np                               # import numpy library as np
import matplotlib.pyplot as plt                  # import mathpltlib.pyplot library as plt
import cmath                                     # import cmath library 

# Shifted Unit step sequence 
N = 10                                         # take N=10 for given sequence y1[n]                                   
n = np.linspace(0,100,101)                     # creating a space of from 0 to 100 i.e. 101 with 101 samples i.e. 1 unit b/w each point
y1 = np.zeros([101])                            # initializing for step signal array
y1[40:2*N+41] = 1                               # Unit step signal between 40 to 60

plt.plot(n,y1,'r')                                  #plot unite step signal
plt.xlabel("n")
plt.ylabel("y[n]")
plt.title("Unit step Signal ") 
plt.figure()


# X[w] = sum + x[n]*exp(-n*w*j)                # where n = -inf to +inf
sumn = 0
j = cmath.sqrt(-1)                             # iota as sqet(-1) using cmath library   
w = np.linspace(-np.pi, np.pi, 101)            # angular frequency in range of - pi to +pi
# w = np.array([0])

Y = np.zeros(len(w), dtype = 'complex')        # create array of zero magnitude for -pi to pi of complex datatpye
shift = (len(n)-1)/2                           # shifting to mid                           
print("shift",shift)

#DTFT of signal y1[n]
for i in range(0,len(Y)):
    for k in range(0,2*N+1):
        sumn = sumn + y1[np.int(shift-N+k)]*np.exp(-(shift-N+k)*w[i]*j)
    Y[i] = sumn
    sumn = 0
Y = Y*np.exp(j*w*((len(n)-1)/2))
print("for w = ",w[10])
print("The value of Y(w) is =",Y[10])
                
plt.plot(w,Y.real,'b')                             # Real part plot of DTFT
plt.xlabel("w")
plt.ylabel("Re( y[n] )")
plt.title("REAL PART PLOT ") 
plt.figure()
plt.plot(w,Y.imag,'black')                             # Imag part plot of DTFT 
plt.xlabel("w")
plt.ylabel("Im( y[n] )")
plt.title("IMAGINARY PART PLOT ") 
```

```python
#phase and magnitude plot 

phaseY = np.zeros([len(w)])              #define total length of phase 
for i in range(0,len(w)):
    phaseY[i] = cmath.phase(Y[i])        #phase of Y[n] using cmath library.
    
# Magnitude and Phase
plt.plot(w, np.abs(Y),'g')               # green color plot of Abs(Y[n]) vs w
plt.xlabel("w")
plt.ylabel("Mag ( y[n] )")
plt.title("Magnitude PLOT ") 
plt.figure()

plt.plot(w, phaseY*180/np.pi,'orange')   # green color plot of Phase(Y[n]) in degrees vs w 
plt.xlabel("w")
plt.ylabel("Phase ( y[n] )")
plt.title("Phase angle PLOT ") 
```

```python
plt.magnitude_spectrum(Y)
plt.title('Magnitude Spectrum using matplotlib')
plt.figure()
plt.phase_spectrum(Y)
plt.title('Phase Spectrum using matplotlib')
```
