
# Median filter for corrupted signal by impulse noise.

```python
import numpy as np                               #import numpy library as np
import matplotlib.pyplot as plt                  #import mathpltlib.pyplot library as plt
n = np.linspace(0,100,101)                       #creating a space of from 0 to 100 i.e. 101 with 101 samples i.e. 1 unit b/w each point
print(len(n))                                    #check length of n

#generate x[n]
s = 2*n*0.9**n                                   #Generating s[n] as given in question 
plt.plot(n,s)                                    #plot s[n] vs n plot
plt.xlabel("n")
plt.ylabel("S[n]")
plt.title("Signal S[n]") 
plt.figure()

#generate random impulse signal as a noise.
impulse_noise = np.zeros([101])                            #define 0 at all the instances 
indices_len = np.random.randint(2,len(impulse_noise)/10)   #number of impulse to be get within that span 

indices = np.zeros([indices_len])                            
for i in range(indices_len):
    indices[i] = np.random.randint(0,len(impulse_noise))

for i in range(indices_len):
    index = np.int(indices[i])
    impulse_noise[index] =  np.random.randint(-1000,1000)   #random impulse of magnitude from -1000 to +1000
    print("index",index)                                    #index whare impulse is generated.

#add randome impulses with original x[n]
x = impulse_noise + s                                 #adding impulse noise to uncorrupted signal 
plt.plot(n,impulse_noise)                             #to plot impulse noise vs n 
plt.xlabel("n")
plt.ylabel("impulse noise")
plt.title("IMPULSE NOISE") 
plt.figure()

k = 2
m= 2*k + 1                                            #window size

v = np.zeros([101])                                   #initialize v[n] as zero throughout length on n 
t = np.zeros([m])
v[0]=x[0]

# median filter to corrupted signal 
for i in range(1,len(x)-k): 
    t = np.sort(x[i:m+i])          
    v[i] = t[np.int(len(t)/2)]
    
plt.plot(n,v)                                        #plot smothern output signal y[n] 
plt.xlabel("n")
plt.ylabel("v[n]")
plt.title("Output Y[n]") 
```
